```python
import json

with open("patient.json", "r") as file:
    data = json.load(file)
```

```python
print(data)
```

```python
official_name = next(n for n in data["name"] if n.get("use") == "official")

full_name = " ".join(official_name.get("given", []) + [official_name.get("family", "")])

print(full_name)
```

```python
print(data.get("gender"))
```

```python
print(data.get("birthDate"))
```

```python
address = data["address"][0]

print("City:", address.get("city"))
print("State:", address.get("state"))
print("Postal Code:", address.get("postalCode"))
```

```python
phones = [t for t in data.get("telecom", []) if t.get("system") == "phone"]

print("Number of phone numbers:", len(phones))
```

```python
types = [t.get("use") for t in phones]

print("Phone types:", types)
```
